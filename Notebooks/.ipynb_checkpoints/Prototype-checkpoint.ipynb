```python
from math import floor, ceil

def simulate_ray(image, n, m, start, end, draw=True):
    """
    Simulates a ray on an image im of shape (n, m)
    starting from start = (row_start, col_start)
    ending in end = (row_end, col_end).
    """
    
    def tiempo_en(densidad):
        return densidad + 1

    def eval_pol(a, b, x):
        return a * x + b
    

    rows, cols = image.shape
    delta_rows = rows / n
    delta_cols = cols / m
    x = np.array([start[0], end[0]])
    y = np.array([start[1], end[1]])
    z = np.polyfit(x, y, 1)
    z_inv = np.polyfit(y, x, 1)
    
    D = np.zeros((n, m))
    
    if draw:
        b, a = z
        x = np.linspace(start[0], end[0], 100)
        y = np.array([a + e*b for e in x])
        plt.plot(x, y, 'r', '-')
        plt.imshow(np.uint8(image))
    
    # filas en las que esta el rayo
    i1 = eval_pol(z[0], z[1], 1) 
    i2 = eval_pol(z[0], z[1], cols)
    
    i_min = max(1, min(rows-1, ceil(min(i1,i2))))
    i_max = max(1, min(rows-1, ceil(max(i1,i2))))
    t = 0
    for i in range(i_min, i_max):
        j1 = eval_pol(z[0], z[1], i-1)
        j2 = eval_pol(z[0], z[1], i)
        j_min = max(1, min(cols, ceil(min(j1, j2))));
        j_max = max(1, min(cols, ceil(max(j1, j2))));
        
        for j in range(j_min, j_max):
            t = t + tiempo_en(float(im[i,j]))
            n_i = int(min(n-1, np.fix(i/delta_rows) + 1))
            m_j = int(min(m-1, np.fix(j/delta_cols) + 1))
            D[n_i, m_j] = D[n_i, m_j] + 1
            
    return D
            
simulate_ray(im, 100, 100, (1, 1), (400, 600), draw=False)
```

Definiremos nuestro modelo de la siguente manera. Dada una imagen de dimension $n \times m$ definiremos la siguiente circunferencia $\mathcal{C}$ centrada en el punto $c = (\frac{n}{2}, \frac{m}{2})$ y de radio $r = \frac{\sqrt{n^2 + m^2}}{2}$.

Luego podemos parametrizar a cualquier recta que pasa por la imagen definiendo cuales son los puntos que atravieza en $\mathcal{C}$.

Ahora que sabemos simular un rayo a partir de los angulos $\alpha$ y $\beta$, podemos simular un conjunto de ellos e intentar recuperar la imagen
